```
# 1. Instalação das ferramentas necessárias
!pip install -q transformers accelerate

import torch
from transformers import pipeline
import psutil
import time

print("🚀 Inicializando Assistente Local Dell (Sem necessidade de API Key)...")

# 2. Configuração do Modelo (Roda direto no Hardware)
try:
    # Detecta se o Colab te deu uma GPU (fica muito mais rápido)
    device = 0 if torch.cuda.is_available() else -1

    # Carrega um modelo potente e leve
    pipe = pipeline(
        "text-generation",
        model="Qwen/Qwen2.5-0.5B-Instruct",
        device=device,
        torch_dtype="auto"
    )
    print("✅ Modelo carregado com sucesso no hardware local!")
except Exception as e:
    print(f"❌ Erro ao carregar modelo: {e}")

# 3. Base de Conhecimento Dell (Seu diferencial)
DELL_ERRORS = {
    "2 bips": "Falha de Memória RAM. Ação: Limpar contatos ou trocar pentes.",
    "5 bips": "Falha na bateria CMOS (Relógio).",
    "7 bips": "Falha no Processador (CPU)."
}

def assistente_senior_local(pergunta):
    start = time.time()

    # Checa erros conhecidos primeiro
    diagnostico = "Análise geral de hardware."
    for erro, solucao in DELL_ERRORS.items():
        if erro in pergunta:
            diagnostico = solucao

    prompt = f"Você é um técnico Dell. Use este diagnóstico: {diagnostico}. Pergunta: {pergunta}"

    # Gera a resposta sem usar internet/API
    res = pipe(prompt, max_new_tokens=200, do_sample=True, temperature=0.7)

    fim = time.time()
    print(f"\n⏱️ Processado localmente em {fim-start:.2f}s")
    return res[0]['generated_text']

# --- TESTE AGORA ---
ticket = "Notebook Dell com 2 bips"
print(assistente_senior_local(ticket))
```
