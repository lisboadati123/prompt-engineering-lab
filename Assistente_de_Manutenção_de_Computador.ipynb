```
# -*- coding: utf-8 -*-
"""
Assistente de Manutenção de Computador com Gemini e Busca do Google
Este script cria um chatbot simples para ajudar com problemas de manutenção em
computadores com Ubuntu Linux ou Windows. Ele usa a API do Gemini para gerar
respostas e a ferramenta de busca do Google para encontrar informações
atualizadas na web.
"""

# Importa as bibliotecas necessárias
import google.generativeai as genai
import os
import json

# **Configuração da Chave de API**
# A sua chave de API é AIzaSyB2ieQ12tPfskRCnGeMNhxs3zpJSTNfNIk.
# Ela é necessária para usar a API do Gemini.
# A função 'configure' já está com sua chave.
genai.configure(api_key="AIzaSyB2ieQ12tPfskRCnGeMNhxs3zpJSTNfNIk")

# Função para fazer a busca na web
# Esta função simula a chamada para a ferramenta de busca do Google.
# Em um ambiente real, esta função faria uma chamada de API para o Google Search.
def search_tool(query):
    """
    Simula uma busca na web e retorna resultados formatados.
    Em um ambiente de execução real, esta função seria uma chamada de API.
    """
    print(f"Buscando na web por: '{query}'...")
    # Substitua este código por uma chamada real à API de busca do Google,
    # usando a ferramenta `google_search` do SDK, se disponível.
    # Exemplo:
    # search_results = google_search.search(queries=[query])
    # return json.dumps([result.to_dict() for result in search_results])

    # Para este exemplo, estamos simulando a busca com uma resposta fixa
    # para que o código seja executável imediatamente.

    # Para um uso real, descomente as linhas acima e remova as linhas abaixo.
    return json.dumps([
        {"title": "Exemplo de Título 1", "snippet": f"Resultado de busca para '{query}'. A solução é..."},
        {"title": "Exemplo de Título 2", "snippet": f"Outro resultado para '{query}'. Recomenda-se fazer..."}
    ])

def main():
    """
    Função principal do chatbot de manutenção.
    """
    print("Olá! Eu sou o seu assistente de manutenção de computador.")
    print("Estou aqui para ajudar com problemas no Ubuntu Linux ou Windows.")

    # Loop principal do chatbot
    while True:
        # Pede ao usuário o sistema operacional
        os_choice = input("\nEm qual sistema operacional você precisa de ajuda (Ubuntu/Windows)? Digite 'sair' para encerrar: ").strip().lower()
        if os_choice == 'sair':
            print("Até logo!")
            break
        if os_choice not in ['ubuntu', 'windows']:
            print("Sistema operacional inválido. Por favor, escolha Ubuntu ou Windows.")
            continue

        # Pede ao usuário o problema
        problema = input(f"Ótimo. Qual o seu problema com o {os_choice.capitalize()}? ").strip()

        # Cria a consulta de busca na web
        query = f"{problema} {os_choice} manutenção de computador"

        # Faz a busca na web
        search_results_json = search_tool(query)
        search_results = json.loads(search_results_json)

        # Prepara o prompt para o Gemini, incluindo os resultados da busca
        prompt = (
            f"Você é um especialista em manutenção de computadores para {os_choice.capitalize()}. "
            "Responda à seguinte pergunta com base em seu conhecimento e nos resultados de busca fornecidos.\n\n"
            f"Pergunta do usuário: '{problema}'\n\n"
            f"Resultados de busca:\n"
            f"{search_results_json}\n\n"
            "Formate sua resposta de forma clara e passo a passo, como se fosse um guia de solução de problemas."
        )

        # Envia o prompt para o modelo do Gemini
        print("\nProcessando... Gerando uma resposta detalhada para você.")

        try:
            modelo = genai.GenerativeModel('gemini-1.5-flash')
            response = modelo.generate_content(prompt)
            print("\n--- Resposta do Assistente ---")
            print(response.text)
            print("----------------------------")
        except Exception as e:
            print(f"Ocorreu um erro ao chamar a API do Gemini: {e}")
            print("Por favor, verifique se sua chave de API e a API de Linguagem Generativa estão ativadas no seu projeto.")

if __name__ == "__main__":
    main()
```
